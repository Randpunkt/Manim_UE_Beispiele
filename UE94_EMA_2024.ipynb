```python
from manim import *
from numpy import *

config.media_width = "75%"
config.verbosity = "WARNING"
```

```python
%%manim -qm Tgt_Circ

class Tgt_Circ(Scene):
    def construct(self):
        
        a = 2 # Radius des Kreieses
        b = 6 # Abstand Punkt P und Mittelpunkt A
        
        
        alpha=pi

        #--------------------------------------------------
        # Angabe
        #--------------------------------------------------
        
        A = array([-a,-3.1,0])        #Mittelpunkt des Kreieses
        P = array([b-a,-3.1,0])
        N = .5*(A + P)
        X = get_intersection(A[0],A[1],a,N[0],N[1],b/2)

        #--------------------------------------------------
        # Punkte und Kreise definieren (mObjects)
        #--------------------------------------------------
        
        pt_size=(a+b)/200
        pt_A = Dot(point=A, radius=pt_size)        #Eckpunkte mObjects -> Dots
        pt_P = Dot(point=P, radius=pt_size)
        pt_N = Dot(point=N, radius=pt_size)
        pt_X = Dot(point=X, radius=pt_size)
        circ_a = Circle(radius = a) #Kreis mit Mittelpunkt A
        circ_a.move_to(A)

        line_a = Line(A,P)
        arc_b = Arc(radius=b/2,angle=alpha, start_angle=0,arc_center=N,color=BLUE)  # Thaleskreis

        #--------------------------------------------------
        # Beschrifrtung der Punkte
        #--------------------------------------------------
        
        label_A = Text("M")              
        label_A.next_to(pt_A, DOWN, buff=0.3)
        label_P = Text("P")              
        label_P.next_to(pt_P, DOWN, buff=0.3)
        label_N = Text("N")              
        label_N.next_to(pt_N, DOWN, buff=0.3)

        #--------------------------------------------------
        # alles skalieren
        #--------------------------------------------------
        
        Kreis = VGroup(pt_A,pt_P,pt_N,label_A,label_P,label_N,circ_a,arc_b,line_a)
        scale_x = config.frame_width/Kreis.width
        scale_y = config.frame_height/Kreis.height
        
        scale = min([scale_x, scale_y])
        
        Kreis = Kreis.scale(scale*0.85)
        Kreis.move_to([0,0,0])

        #--------------------------------------------------
        # Animation
        #--------------------------------------------------

        

        
        self.play(FadeIn(pt_A))
        self.play(Create(circ_a))
        self.wait(.5)
        self.play(FadeIn(pt_P))
        self.play(Write(label_A),Write(label_P))
        self.wait(.5)
        self.play(Create(line_a))
        self.wait(.5)
        self.play(FadeIn(pt_N))
        self.play(Write(label_N))
        self.wait(.5)        
        self.play(Create(arc_b))
        self.play(FadeIn(pt_X))
        #self.play(FadeIn(pt_C))
        #self.play(Write(label_C))
        self.wait()
        #self.play(Create(line_a),Create(line_b))
        #self.wait(1.5)


#def get_intersection(M1,r1,M2,r2):
#    [x1,y1] = M1
#    [x2,y2] = M2
#
#    d = ((x1-x2)**2+(y1-y2)**2)**(1/2)
#    a = (r1**2-r2**2+d**2)/(2*d)
#    h = (r1**2-a**2)**(1/2)
#    P2 = M1+a/d*(M2-M1)
#    x3 = P[0]
#    y3 = P[1]
#    x4 = x3 + h/d * (y2-y1)
#    y4 = y3 + h/d * (x2-x1)#
#
#    return (x4,y4)


def get_intersections(x0, y0, r0, x1, y1, r1):
    # circle 1: (x0, y0), radius r0
    # circle 2: (x1, y1), radius r1

    d=math.sqrt((x1-x0)**2 + (y1-y0)**2)
    
    # non intersecting
    if d > r0 + r1 :
        return None
    # One circle within other
    if d < abs(r0-r1):
        return None
    # coincident circles
    if d == 0 and r0 == r1:
        return None
    else:
        a=(r0**2-r1**2+d**2)/(2*d)
        h=math.sqrt(r0**2-a**2)
        x2=x0+a*(x1-x0)/d   
        y2=y0+a*(y1-y0)/d   
        x3=x2+h*(y1-y0)/d     
        y3=y2-h*(x1-x0)/d 

        x4=x2-h*(y1-y0)/d
        y4=y2+h*(x1-x0)/d
        
        return (x3, y3, x4, y4)
```
